## 데이터 프레임 형식 맞추기 및 병합

```python
import pandas as pd

def process_date(df: pd.DataFrame, date_column: str, current_format: str) -> pd.DataFrame:
    """
    날짜 형식을 표준 형식(YYYYMMDD)으로 변환하는 함수

    매개변수:
        df (pd.DataFrame): 날짜 형식을 변환할 데이터프레임
        date_column (str): 변환할 날짜 열의 이름
        current_format (str): 현재 날짜 형식

    반환값:
        pd.DataFrame: 날짜 형식이 변환된 데이터프레임
    """
    # 날짜 열을 문자열로 변환하고 ".pdf" 확장자 제거
    df[date_column] = df[date_column].astype(str).str.replace('.pdf', '', regex=False)

    # 이미 YYYYMMDD 형식인지 확인 후 변환
    if not df[date_column].str.match(r'^\d{8}$').all():
        df[date_column] = pd.to_datetime(df[date_column], format=current_format).dt.strftime('%Y%m%d')

    return df

def combine_dataframes(dfs: list) -> pd.DataFrame:
    """
    여러 데이터프레임을 하나로 합치는 함수

    매개변수:
        dfs (list): 합칠 데이터프레임들의 리스트

    반환값:
        pd.DataFrame: 합쳐진 데이터프레임
    """
    return pd.concat(dfs, ignore_index=True)

def main():
    # 각 데이터프레임에 대해 날짜 형식 처리 수행
    df_han = process_date(df_han, 'date', '%Y.%m.%d')
    df_edaily = process_date(df_edaily, 'date', '%Y-%m-%d')
    df_mk = process_date(df_mk, 'date', '%Y-%m-%d')
    df_bond = process_date(df_bond, 'date', '%Y-%m-%d')
    df_call_rate = process_date(df_call_rate, 'date', '%Y.%m.%d')

    # 데이터프레임 합치기, 위원회 결정보고서 제외
    df_combined = combine_dataframes([df_han, df_edaily, df_mk, df_bond, df_call_rate])

    # 결과 확인
    print(df_combined.head())

if __name__ == "__main__":
    main()
```

```python
# df_combined의 총 길이 확인
length = len(df_combined)
print(f"df_combined의 총 길이: {length}")
```

```python
# 날짜 형식이 YYYYMMDD가 아닌 행을 골라내기
df_combined['date'] = df_combined['date'].astype(str)  # Ensure date column is string
incorrect_date_format = ~df_combined['date'].str.match(r'^\d{8}$')

# 날짜 형식이 잘못된 데이터프레임
df_incorrect_dates = df_combined[incorrect_date_format]

# 결과 확인
print(f"형식이 잘못된 날짜를 가진 행의 수: {len(df_incorrect_dates)}")
print(df_incorrect_dates.head())
```

```python
# 중복된 date 값 처리
df_call_rate = df_call_rate.drop_duplicates(subset='date', keep='first')


# df_combined의 date 열을 문자열로 변환
df_combined['date'] = df_combined['date'].astype(str)

# df_call_rate의 date 열을 문자열로 변환
df_call_rate['date'] = df_call_rate['date'].astype(str)

# 데이터프레임 합치기
df_combined = pd.concat([df_han, df_edaily, df_mk, df_bond], ignore_index=True)

# 날짜를 기준으로 정렬
df_combined.sort_values(by='date', inplace=True)

# processed_call_rate의 rate_change 값을 df_combined에 추가
df_combined = df_combined.merge(df_call_rate[['date', 'rate_change']], on='date', how='left')

# 결과 확인
print(f"df_combined의 총 길이: {len(df_combined)}")
print(df_combined.head())
```

```python
# 2005년 06월 09일 이후 데이터만 사용합니다.

# 'date' 열을 다시 날짜 형식으로 변환 (YYYYMMDD -> datetime)
df_combined['date'] = pd.to_datetime(df_combined['date'], format='%Y%m%d')

# 원하는 날짜 이후의 데이터만 필터링
df_filtered = df_combined[df_combined['date'] >= '2005-06-09']

# 날짜 형식을 다시 YYYY-MM-DD -> YYYYMMDD로 변환
df_combined = process_date(df_filtered, 'date', '%Y-%m-%d')

# 인덱스 초기화
df_combined = df_combined.reset_index(drop=True)

# 결과 확인
print(df_combined.head())
```
